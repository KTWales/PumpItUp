# Pump It Up: Data Mining the Water Table

**Can you predict which water pumps are faulty?**

Using data from Taarifa and the Tanzanian Ministry of Water, can you predict which pumps are functional, which need some repairs, and which don't work at all?

Predict one of these three classes based on a number of variables about what kind of pump is operating, when it was installed, and how it is managed.

A smart understanding of which waterpoints will fail can improve maintenance operations and ensure that clean, potable water is available to communities across Tanzania.

_Website Link:_ https://www.drivendata.org/competitions/7/pump-it-up-data-mining-the-water-table/

### Feature Engineering

Selecting and renaming variables.
This is to reduce the number of independent variables in the statistical model.

```python
# List of column names
train_columns = columns[:]
train_columns
```

```python
# No. of Unique values in features
for i in train_columns:
  n = train[i].unique()
  if len(n) < 20:
    print(i, " : ", train[i].nunique(), " Volume count")
    print(i, " : ", n, " Unique values")
  else:
    print(i, " : ", str(len(n)), " Unique values")
```

```python
# Regrouping specific independent variables
train['district_code_recoded'] = np.where(train['district_code'] <= 4, 'District Codes 1-4', 'Other Districts')
train['scheme_management_recoded'] = np.where(train['scheme_management'] == 'VMC', 'VMC', 'Other')
train['extraction_type_recoded'] = np.where(train['extraction_type'] == 'gravity', 'gravity', 'other')
train['management_recoded'] = np.where(train['management'] == 'vmc', 'vmc', 'other')
train['management_group_recoded'] = np.where(train['management_group'] == 'user-group', 'user-group', 'other')

train['payment_recoded'] = np.where(train['payment'] == 'never pay', 'never pay', 'other') #repetition ??
train['payment_type_recoded'] = np.where(train['payment_type'] == 'never pay', 'never pay', 'other') #repetition ??

train['quality_recoded'] = np.where(train['water_quality'] == 'soft', 'soft', 'other') #repetition ??
train['quality_group_recoded'] = np.where(train['quality_group'] == 'good', 'good', 'other') #repetition ??

train['source_recoded'] = np.where(~train['source'].isin(['shallow well','spring']), 'other', train['source']) #repetition ??
train['source_type_recoded'] = np.where(~train['source_type'].isin(['shallow well','spring','borehole','river/lake']), 'other', train['source_type']) #repetition ??

train['waterpoint_type_recoded'] = np.where(~train['waterpoint_type'].isin(['communal standpipe','hand pump']), 'other', train['waterpoint_type']) #repetition ??
train['waterpoint_type_group_recoded'] = np.where(~train['waterpoint_type_group'].isin(['communal standpipe','hand pump']), 'other', train['waterpoint_type_group']) #repetition ??

train['quantity_recoded'] = np.where(~train['quantity'].isin(['enough','insufficient']), 'other', train['quantity']) #repetition ??
train['quantity_group_recoded'] = np.where(~train['quantity_group'].isin(['enough','insufficient']), 'other', train['quantity_group']) #repetition ??

train['source_class_recoded'] = np.where(~train['source_class'].isin(['groundwater','surface']), 'other', train['source_class'])
```

```python
# Recoded and clean Dataset
train_clean = train[['amount_tsh',
 'gps_height',
 'num_private',
 'basin',
 'region',
 'region_code',
 'district_code_recoded',
 'population',
 'public_meeting',
 'recorded_by',
 'scheme_management_recoded',
 'permit',
 'construction_year',
 'extraction_type_recoded',
 'extraction_type_group',
 'extraction_type_class',
 'management_recoded',
 'management_group_recoded',
 'payment_recoded',
 'quality_recoded',
 'quality_group_recoded',
 'source_recoded',
 'source_type_recoded',
 'source_class_recoded',
 'waterpoint_type_recoded',
 'waterpoint_type_group_recoded',
 'status_group',
 'payment_type_recoded',
 'quantity_recoded',
 'quantity_group_recoded']]


columns = ['amount_tsh','gps_height','num_private','basin','region','region_code',
           'district_code_recoded','population','public_meeting','recorded_by','scheme_management_recoded','permit','construction_year',
           'extraction_type_recoded','extraction_type_group','extraction_type_class','management_recoded','management_group_recoded','payment_recoded','quality_recoded','quality_group_recoded',
           'source_recoded','source_type_recoded','waterpoint_type_recoded','waterpoint_type_group_recoded','status_group','payment_type_recoded','quantity_recoded','quantity_group_recoded','source_class_recoded']

categorical = ['basin','region','region_code','district_code_recoded','public_meeting',
               'recorded_by','scheme_management_recoded','permit','extraction_type_group',
               'extraction_type_class','management_recoded','management_group_recoded','payment_recoded','payment_type_recoded','quality_recoded','quality_group_recoded','quantity_recoded',
               'quantity_group_recoded','source_recoded','source_type_recoded','source_class_recoded','waterpoint_type_group_recoded','status_group','extraction_type_recoded', 'waterpoint_type_recoded']
```

```python
table = TableOne(train_clean, 
                 columns = columns, 
                 categorical = categorical,
                #  groupby = groupby, 
                 nonnormal = nonnormal,
                 labels = labels,
                 groupby=['status_group'],
                 pval=True)
print(table.tabulate(tablefmt = "fancy_grid"))
#status_group = f(everything else)
```
